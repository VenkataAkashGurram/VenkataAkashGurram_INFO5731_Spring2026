## Question 2 ( 2 point)
Create a Python program that interactively takes two numerical inputs from the user and swaps their values without utilizing an extra variable.

```python
# Take two numerical inputs from the user
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

# Display values before swapping
print("Before swapping:")
print("a =", a)
print("b =", b)

# Swap the values without using an extra variable
a, b = b, a

# Display values after swapping
print("After swapping:")
print("a =", a)
print("b =", b)


```

## Question 4 ( 2 point)
Design a Python function to check if a number is prime.


```python
import math

def is_prime(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False

    # 2 is the only even prime number
    if n == 2:
        return True

    # Eliminate all other even numbers
    if n % 2 == 0:
        return False

    # Check for factors from 3 up to the square root of n
    # We skip even numbers in the loop by using a step of 2
    limit = int(math.sqrt(n)) + 1
    for i in range(3, limit, 2):
        if n % i == 0:
            return False

    return True

# Testing the function
number = int(input("Enter a number to check: "))
if is_prime(number):
    print(f"{number} is a prime number!")
else:
    print(f"{number} is not a prime number.")

```

## Question 5 ( 2 point)
Write a Python program to concatenate following dictionaries to create a new one.

Sample Dictionary : dic1={1:10, 2:20}, dic2={3:30, 4:40}, dic3={5:50,6:60},

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python
# Sample Dictionaries
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

# Method 1: Using the Unpacking Operator (**)
# This creates a new dictionary by expanding the contents of the others
new_dict = {**dic1, **dic2, **dic3}

# Displaying the result
print(f"Concatenated Dictionary: {new_dict}")
```
